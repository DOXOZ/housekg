```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
soup_list = []

for i in range(100):
    url = f"https://www.house.kg/kupit-kvartiru?page={i}"
    html_text = requests.get(url).text
    
    soup = BeautifulSoup(html_text, 'html.parser')
    
    soup_list.append(soup)

    time.sleep(1)
```

```python
apartment = soup.find('div', class_ = 'listing')
```

```python
apart = []
address = []
price = []
price_kgs = []
price_usd = []
number_of_rooms = []
area = []
floor = []
max_floor = []
for it in soup_list:
    apartments = it.find_all('div', class_ = 'listing')
    for item in apartments:
        apart.append(item.find('p', class_ = 'title').text.replace(' ', '').replace('\n','').split(','))
        address.append(item.find('div', class_ = 'address').text.replace(' ', '').replace('\n',''))
        price.append(item.find('div', class_ = 'sep main'))
    for i in apart:
        number_of_rooms.append(i[0][0])
        area.append(i[1][:-2])
        floor.append(i[2].replace("этажиз", " ").split()[0])
    for i in price:
        price_usd.append(i.find('div', class_='price').text[1:].replace(' ', ''))
        price_kgs.append(i.find('div', class_='price-addition').text[:-3].replace(" ", '' ))
```

```python
df = pd.DataFrame(list(zip(number_of_rooms,area,floor, address, price_usd,price_kgs)),
                  columns = ['Количество комнат', 'Площадь', 'Этаж', 'Адрес', 'Цена в usd', 'Цена в kgs'])
```

```python
columns = ['Количество комнат', 'Площадь', 'Этаж', 'Цена в usd', 'Цена в kgs']
for i in columns:
    df[i] = df[i].astype(float)
```

```python
df.to_csv("Apartments.csv")
```
